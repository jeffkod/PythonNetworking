```python
for device in devices:
    router = Router(ip=device['ipaddress'],username=device['username'],password=device['password'],secret=device['secret'])
    hostname_from_dev = router.get_hostname_from_dev()
    if hostname_from_dev == device['hostname']:
        print("Hostname verified for ",hostname_from_dev)
    else:
        print("Hostnames not matching!!!",hostname_from_dev,device['hostname'])
    device.update(hostname_actual=hostname_from_dev)
#     print(router.get_hostname_from_dev())
#     print(router.show_int_desc())
```
